# Stellar death

For scale:
- Solar mass: $M_☉ = 2.0 \times 10^{30} kg$
- Jupiter mass: $M_J = 1.9 \times 10^{27} kg$

## Basics

- Gases aggregate due to gravity -> increases temperature and pressure -> start fusing elements
- Hydrogen fuses into helium, helium into oxygen and carbon, carbon and oxygen into neon, iron, ...
- Stars can sustain several fusions simultaneously, with a series of concentric shells (heavier elements in the center, hydrogen on the outside)
- Equilibrium between gravity pressure (collapsing) and nuclear fusion + degeneracy pressure (ie. Pauli exclusion principle)

## Birth

Stars form from the gravitational collapse of (mostly hydrogen) gas clouds.

### Brown dwarves
Some protostars never reach temperatures high enough for nuclear fusion, they are **brown dwarves**.

An object is a brown dwarf if:
- it is massive enough to fuse deuterium ($m > 2.5 \times 10^{28} kg = 13 M_J = 0.0125 M_☉$)
- it is not massive enought to fuse hydrogen ($m < 1.6 \times 10^{29} kg = 0.08 M_☉$)

If it's smaller, it's a *sub-brown dwarf* or a planet.

### Main sequence

Heavier objects eventually reach a temperature of 10 million kelvin and start hydrogen fusion (-> deuterium -> helium).

In this state, the energy released by the core of the star (from the fusion) balances the gravitational pressure and stops the gravitational collapse. The star has reached a stable state: the *main sequence*.

- Small, cold, red stars fuse hydrogen slowly and remain on the main sequence for hundreds of billions of years.
- Massive, hot stars can leave the main sequence after a "just" a few million years.

## Death

When the core exhausts its supply of hydrogen the outward pressure generated by fusion is not enough to counteract the gravity and the star (and the core) contracts. The star starts to evolve off the main sequence. One of two things happens next:
- the core becomes hot enough to fuse helium
- [electron degeneracy pressure](https://en.wikipedia.org/wiki/Electron_degeneracy_pressure) (electrons being smashed too close to each other -> Pauli exclusion principle -> repulsion) becomes sufficient to stop the collapse

- **Red dwarfs ($m < 0.5 M_☉$)**: fuses all hydrogen into helium then turn into white dwarves. Theoretical because such stars live much longer than the age of the universe.
- **Red giants**:
  - $0.5 M_☉ < m < 1.5 M_☉$: expands into red giant then turns into a white dwarf.
  - $1.5 M_☉ < m < 8 M_☉$: expands into red giant -> not massive enugh for carbon fusion -> contraction -> superwinds produces a *nebula* + extremely hot central star which cools into a white dwarf
- **Supergiants**: helium ignition starts before Electron Degeneracy Pressure
  - $8 M_☉ < m < 40 M_☉$: fuse elements all the way to nickel and iron (inert) -> less outward pressure -> rapid collapse of the star -> *type II supernova*
  - $40 M_☉ < m$: radiation pressure is so high that they lose mass before they become supergiants
    - $m_{core} < 1.4 M_☉$: turns into a white dwarf + planetary nebula
    - $2.5 M_☉ < m_{core}$: core gets hot enough to break down neon and starts a fusion chain reaction to iron -> collapse into neutron star or a black hole + type Ib, Ic or II supernova (which fuses all the way to uranium!)

## Remnants

### White dwarfs

A white dwarf has run out of fuel and radiates energy (in the form of light and heat).

Since white dwarfs cannot sustain thermonuclear reactions anymore (and so aren't stars anymore), the only remaining barrier to gravity (and complete collapse) is the degeneracy pressure of the electrons.
Because of this, they are relatively small (for a star of $1 M_☉$, the mass of the white dwarf would be around $0.6 M_☉$ for approximately the volume of the Earth).

- Higher mass white dwarfs have a smaller volume (because more gravity).
- The composition of a white dwarf depends on it's mass: larger stars fuse more elements (helium, oxygen, carbon, magnesium).
- White dwarfs heavier than the Chandrasekhar limit ($1.4 M_☉$) may overcome electron degeneracy pressure and start a runaway nuclear reaction leading to a type IA supernova.
- In a binary system hydrogen from the larger companion may create a runaway reaction on the white dwarf's surface (a nova).

### Black dwarfs

Once a white dwarf has radiated all of its energy (or enough to no longer emit light), it becomes a black dwarf.
Black dwarfs don't actually exist yet because the time it would take to reach this state is longer than the age of the universe.

### Neutron stars

They are the collapsed core of a large star (10 to 29 $M_☉$): when such a star collapses the pressure causes protons and electrons to fuse (by [electron capture](https://en.wikipedia.org/wiki/Electron_capture)), creating a *neutron* and an electron neutrino.

Without electrons to keep nuclei apart, neutrons collapse into a dense ball (radius of the order of $10 km$!).
The neutrons resist further gravitational pressure by the Pauli Exclusion Principle (similar to the electron degeneracy pressure, but much sronger).

Neutron stars rotation speed increases as the star shrinks (down to a period of $1.5ms$!).
These stars are called pulsars: they emit radiation at a very regular frequency.

### Black holes

If the mass of a netron star is high enough the neutron degeneracy pressure will be insufficient to prevent collapse below the [Schwarzschild radius](https://en.wikipedia.org/wiki/Schwarzschild_radius): the neutron star thus becomes a black hole.
