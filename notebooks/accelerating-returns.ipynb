# The law of accelerating returns

This note it based on the appendix of [Ray Kurzweil](https://en.wikipedia.org/wiki/Ray_Kurzweil)'s [The Singularity is Near](https://en.wikipedia.org/wiki/The_Singularity_Is_Near).

## Abstract

The *law of accelerating returns* states that evolution (in a broad sense) is a doubly exponential phenomenon. This is especially valid for information-based processes and technologies, including our knowledge of human intelligence.

$$ W = A e^{B e^{t}} $$

Where $W$ is our knowledge of the world, especillay regarding building computational devices.

## Infering the law of accelerating returns

Let's consider four variables:

- $V$: velocity (power) of compuatation (measured in calculations per second per unit cost)
- $W$: world knowledge as it pertains to designing and building computational devices (this is the *evolution* we measure)
- $N$: resources available for computations (in unit cost)
- $t$: time

### First order analysis

Assumptions:

- the computer power ($V$) is a linear function of our knowledge ($W$) (this is a conservative assumption): $ V = c_1 W $
- the resources deployed for computations are growing exponetially: $N = c_3 e^{c_4 t}$
- the rate of change of world knowledge is proportional to the product of the velocity of computation and the deployed resources: $\frac{dW}{dt} = c_2 N V$

Which yields a differential equation on $W$:

$$ \frac{dW}{dt} = c_1 c_2 c_3 e^{c_4 t} W $$

Which gives:

$$ W = W_0 \exp \left( \frac{c_1 c_2 c_3}{c_4} e^{c_4 t} \right) $$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.0, 5.0, 0.01)

w_0 = 1
c = 1
c_4 = 1

W = w_0 * np.exp((c/c_4) * np.exp(c_4 * t))
plt.semilogy(t, W)

plt.xlabel('Time (s)')
plt.ylabel('World knowledge (W)')
plt.grid(True)
```

### Refined analysis

The previous assumption that $V$ is proportional to $W$ is very conservative, in reality innovations have miltiplicative effect rather than additive impact on the ongoing rate:

$$ V = C^W $$

And so $\frac{dW}{dt} = C^W$ and:

$$ W = \frac{1}{\ln C} \ln \left( \frac{1}{1 - t \ln C} \right) $$

Which has a slow logarithmic growth while $ t < 1 / \ln C $ but then explodes (singularity) when t = $ 1 / \ln C $.

```python
t = np.arange(0, 5, 0.01)

C = 1.5

W = (1/np.log(C)) * np.log(1/(1 - t * np.log(C)))
plt.plot(t, W)

plt.xlabel('Time (s)')
plt.ylabel('World knowledge (W)')
plt.grid(True)
```

### Network effect

If the growth rate instead includes a logarithmic network effect, we get:

$$ \frac{dW}{dt} = W + W \ln(W) $$

Which yields:

$$ W = \exp \left( e^t \right) $$

```python
t = np.arange(0, 5, 0.01)

C = 1.5

W = np.exp(np.exp(t))
plt.semilogy(t, W)

plt.xlabel('Time (s)')
plt.ylabel('World knowledge (W)')
plt.grid(True)
```
